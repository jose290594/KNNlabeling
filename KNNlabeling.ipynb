```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Get the data

```python
df = pd.read_csv('jose_knn')
```

**Check the head of the dataframe.**

```python
df.head() 
```

## Plot for visualization

```python
sns.pairplot(df,hue='TARGET',palette='coolwarm')
```

## Standardize the variables

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('TARGET',axis=1))
scaled_features = scaler.transform(df.drop('TARGET',axis=1))
#scaled_features = scaler.fit_transform(df.drop('TARGET CLASS',axis=1))
```

## Create new and scaled dataframe

```python
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat.head()
```

## Split data into training and testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features,df['TARGET'],
                                                    test_size=0.30)
```

## Create and train model for KNN labeling

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

## First predictions round with k=1

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
#confusion matrix and classification report for checking
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

## Choosing a new K value

```python
error_rate = []

#Loop for testing a range of K values and append the error rate for later comparison
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

## Plot for error rate vs K value comparison

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('error rate vs K values')
plt.xlabel('K')
plt.ylabel('error rate')
#The lower value is 10 (balance between less error rate and lower K value)
```

## Retrain with new K value

```python
knn = KNeighborsClassifier(n_neighbors=10)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=10')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```
